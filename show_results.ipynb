```python
import matplotlib.pyplot as plt
```

```python
num_steps = [1, 2, 4, 8, 16, 32, 64, 128, 256]
num_steps_cpu_perf = [0.317407, 0.602152, 1.090124, 2.207250, 4.280052, 8.019920, 15.201359, 31.758209, 83.506626]
num_steps_gpu_perf = [0.001974, 0.003627, 0.005941, 0.012444, 0.031390, 0.058521, 0.114378, 0.227405, 0.594065]

fig, ax1 = plt.subplots()
cpu_line, = ax1.plot(num_steps, num_steps_cpu_perf, color='b', label='CPU Performance')
ax1.set_xlabel('Number of Steps')
ax1.set_ylabel('CPU Performance', color='b')
ax1.tick_params(axis='y', labelcolor='b')

ax2 = ax1.twinx()
gpu_line, = ax2.plot(num_steps, num_steps_gpu_perf, color='r', label='GPU Performance')
ax2.set_ylabel('GPU Performance', color='r')
ax2.tick_params(axis='y', labelcolor='r')

lines = [cpu_line, gpu_line]
labels = [line.get_label() for line in lines]
ax1.legend(lines, labels, loc='upper left')

fig.tight_layout()
plt.title("CPU vs GPU Performance vs. Steps (256 x 256 board)")
plt.show()
```

```python
board_size = [1, 2, 4, 8, 16, 32, 64, 128, 256]
board_size_perf_cpu = [0.006412, 0.015098, 0.051729, 0.188661, 0.885342, 3.782530, 15.134221, 63.837132, 231.605345]
board_size_perf_gpu = [0.473876, 0.530003, 0.479988, 0.504985, 0.557967, 0.637970, 0.589587, 0.524172, 0.477212]

plt.plot(board_size, board_size_perf_cpu, label="CPU Performance", c='blue')
plt.plot(board_size, board_size_perf_gpu, label="GPU Performance", c='red')

plt.legend()
plt.xlabel("Board Size (Side Length)")
plt.ylabel("Execution Time (ms)")
plt.title("Board Size vs. Execution Time for CPU and GPU (256 steps)")
plt.show()
```
